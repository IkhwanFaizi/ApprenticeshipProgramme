```python
# Data Structures
# A data structure is a collection of data values.
# Data Structure can be defined as the group of data elements
# which provides an efficient way of storing and organising data
# in the computer so that it can be used efficiently

# LIST     TUPLE     SETS     DICTIONARIES

# List In Python
# List is a collection of Data Values
# In Python, List is denoted by using []
# List is MUTABLE --> changeable or modifiable
# List allow you to have duplicatw elements

# Create A List

mylist = [10,20,30,40,50,60,70]
print(type(mylist))
print(mylist)

mylist1 = list([100,200,300,400])
print(type(mylist1))

# create an empty list

mylist2 = []
print(type(mylist2)) 
```

```python
mylist = [10,20,30,40,50,60,70]
# indexing
print(mylist[0])
# negative indexing
print(mylist[-1])
# slicing
print(mylist[1:5])
```

```python
# iteration over list

mylist = [10,20,30,40,50,60,70]

for everyValue in mylist:
  print(everyValue)
```

```python
mylist = [10,20,30,40,50,60,70]
print(len(mylist))
```

```python
i = 0
while i<len(mylist):
  print(mylist[i])
  i = i +1
```

```python
# List Is Mutable
mylist = [10,20,30,40,50,60,70]
print(mylist)

mylist[1] = 200
print(mylist)
```

```python
mylist = [10,20,30,40,50,60,70]
print(mylist)

mylist[:3] = [100,200,300]
print(mylist)
```

```python
# List In-Built Methods

mylist = [10,20,30,40,50,60,70,80]
print(mylist)

# 1. append(value) --> used to add an item at the end of your list
mylist.append(90)
print(mylist)

# 2. insert(index,value) --> used to add an item at the specific position
mylist.insert(1,100)
print(mylist)

# 3. pop(index) --> used to delete an item from the specified index
mylist.pop(1)
print(mylist)

# 4. pop() --> if no argument is specified, then it will delete the last element
mylist.pop()
print(mylist)

# 5. remove(value) -- > delete first occurence if there is duplicated calue
mylist.remove(80)
print(mylist)

# 6. reverse() --> used to print List in reverse order
mylist.reverse()
print(mylist)
```

```python
# Tuple
# Tuple is a collection of Data Values
# In Python, Tuple is denoted by using ()
# Tuple is IMMUTABLE

# Create a Tuple

tuple1 = (10,20,30,40,50)
print(type(tuple1))
print(tuple1)

tuple2 = tuple([1,2,3,4,5])
print(type(tuple2))

# empty Tuple
tuple3 = ()
print(type(tuple3))
```

```python
myVar = 10, 20, 30, 90.21, "Wan", True
print(type(myVar))
```

```python
# create tuple having one element only
tup1 = (10,) # put comma so, it not become int type
print(type(tup1))
```

```python
# Tuple is Immutable

tup1 = (10,20,30,40,50)
print(tup1)

tup1[1] = 200
print(tup1)
```

```python
# Tuple In-Built Methoda
tup1 = (10,20,30,40,50,60,70,10,10,10)

print(len(tup1))
print(max(tup1))
print(min(tup1))
# count will return the number of times specified value appears
print(tup1.count(10))
# index will find the first occurence of the specified value
print(tup1.index(50)) 
```

```python
# Set
# Set is an Unordered Collection of Data Values
# Set is denoted by using {}
# Set does Not Allow Duplicate Elements

# Create a Set

myset = {10,20,30,40,50,10,20,30}
print(type(myset))
print(myset)

myset1 = set([10,20,30,40,50])
print(type(myset1))

# Create an empty set ~ {} dict

myset2 = {}
myset3 = set()
print(type(myset2))
print(type(myset3))
```

```python
# WAP that should print number of unique values in LIST

mylist = [10,20,30,10,30,40,50,60,30,70]
empty_list = []

for i in mylist:
  if i not in empty_list:
    empty_list.append(i)
print("Number of Unique Values Are :", len(empty_list))
print(empty_list)
```

```python
mylist = [10,20,30,10,30,40,50,60,30,70]
print("Number of Unique Values Are :", len(set(mylist)))
```

```python
# DICTIONARIES
# Used to store information in key:value pair format
# Dictionary is denoted by using {}

# Create a dictionary

stInfo = {'rno': 123, 'name':'Wan', 'age':23, 'rno':456} #key should unique 
print(type(stInfo))
print(stInfo)

print(stInfo['rno'])
print(stInfo['name'])
print(stInfo.get('age'))
```

```python
# Iteration over Dictionary

for keys in stInfo:
  print(keys)
```

```python
# print dictionary values

for val in stInfo.values():
  print(val)
```

```python
# printing key value together

for k,v in stInfo.items():
  print(k,'-->',v)
```
