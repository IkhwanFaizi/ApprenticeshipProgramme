```python
# Logistic Regression
import pandas as pd
df = pd.read_csv('https://trello-attachments.s3.amazonaws.com/6062c6af283e86843a32b025/60c7f8fa059b3a683ff0f11d/312991420badd1c0264eac0d335e3bbf/diabetes.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
# Extracting Independent and Dependent Variable
feature_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
                'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[feature_cols]  # Independent Variable
y = df.Outcome        # Dependent Variable
```

```python
# Split our dataset into Training Set and into Testing Set
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
# Model Development and Prediction

from sklearn.linear_model import LogisticRegression

# Create Logistic Regression Classifier Object

logreg = LogisticRegression()

# Fit the model with data

logreg.fit(x_train , y_train)
```

```python
# Predictions
y_pred = logreg.predict(x_test)
y_pred
```

```python
# Evaluate Our Model Using Confusion Matrix
# it is a matrix used to evaluate the performance of classification model
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test , y_pred)
cnf_matrix

"""
             Predicted_No     Predicted_Yes

Actual_No     TN = 115           FP = 15


Actual_Yes    FN = 25            TP = 37

"""
```
