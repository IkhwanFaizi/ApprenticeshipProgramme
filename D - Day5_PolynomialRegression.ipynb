```python
# Polynomial Regression
import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd

# importing dataset
data_set = pd.read_csv('https://trello-attachments.s3.amazonaws.com/6062c6af283e86843a32b025/60d0152706cf05391bac797c/164d4521d23ff95a2989c5f66e08ff33/Position_Salaries.csv')
data_set.head()
```

```python
 # Extracting Independent And Dependent Variable
x = data_set.Level.values[:,nm.newaxis]
y = data_set.Salary.values
```

```python
x
```

```python
y
```

```python
# Building The Linear Regression Model
# Fitting The Linear regression to the dataset

from sklearn.linear_model import LinearRegression
lin_regs = LinearRegression()
lin_regs.fit(x,y)
```

```python
# Fitting The Polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_regs = PolynomialFeatures(degree=2)
x_poly = poly_regs.fit_transform(x)
x_poly

# when degree is set to 2 the features created will be  1 , x1 ,  x2
# when degree is set to 3 the features created will be  1 , x1 ,  x2 , x3
# when degree is set to 4 the features created will be  1 , x1  , x2 , x3 , x4
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly , y)
```

```python
# Visualizing the result for Linear Regression Model
mtp.scatter(x,y,color="blue")
mtp.plot(x,lin_regs.predict(x),color="red")
mtp.title('Detection Model Using Linear Regression')
mtp.xlabel('Position Levels')
mtp.ylabel('Salary')
mtp.show() 
```

```python
# Visualize The Result For Polynomial Regression
mtp.scatter(x,y,color="blue")
mtp.plot(x,lin_reg_2.predict(poly_regs.fit_transform(x)),color="red")
mtp.title('Detection Model Using Polynomial Regression (Degree = 2)')
mtp.xlabel('Position Levels')
mtp.ylabel('Salaries')
mtp.show()
```

```python
# For Degree = 3

poly_regs = PolynomialFeatures(degree=3)
x_poly = poly_regs.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly , y)
```

```python
# Visualizing the result for Polynomial Regression Model when degree = 3
mtp.scatter(x,y,color="blue")
mtp.plot(x,lin_reg_2.predict(poly_regs.fit_transform(x)),color="red")
mtp.title('Detection Model Using Polynomial Regression (Degree = 3)')
mtp.xlabel('Position Levels')
mtp.ylabel('Salaries')
mtp.show()
```

```python
# For Degree = 4

poly_regs = PolynomialFeatures(degree=4)
x_poly = poly_regs.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly , y)
```

```python
# Visualizing the result for Polynomial Regression Model when degree = 4
mtp.scatter(x,y,color="blue")
mtp.plot(x,lin_reg_2.predict(poly_regs.fit_transform(x)),color="red")
mtp.title('Detection Model Using Polynomial Regression (Degree = 4)')
mtp.xlabel('Position Levels')
mtp.ylabel('Salaries')
mtp.show()
```

```python
# Predicting The Final Output Using Linear Regression Model

lin_pred = lin_regs.predict([[6.5]])
print(lin_pred)
```

```python
# Predicting The Final Output Using Polynomial Regression Model
 
poly_pred = lin_reg_2.predict(poly_regs.fit_transform([[6.5]]))
print(poly_pred)
```
