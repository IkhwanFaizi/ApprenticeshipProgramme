```python
# Matplotlib

import matplotlib.pyplot as plt
# plot([x-axis],[y-axis])
plt.plot([1,2,3],[4,5,1])
# show() --> used to display the graph
plt.show()
```

```python
# add titles and labels to a graph
x = [5,8,10]
y = [12,16,6]
plt.plot(x,y)
plt.title('Info')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

```python
# Add style to a graph
from matplotlib import style
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'g',label='Line One', linewidth=5) # 'g' = green
plt.plot(x2,y2,'c', label='Line Two', linewidth=5)
plt.title('Info')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend()
plt.grid(True,color='k') # k --> black color
plt.show()
```

```python
# Generate A Bar Graph
plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Example One')
plt.bar([2,4,6,8,10],[8,6,2,5,6],label='Example Two')
plt.legend()
plt.title('Bar Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

```python
# scatter plot
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]
plt.scatter(x,y,label='scatter',color='g',s=25,marker='o') # 's' = size
plt.title('Scatter Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend()
plt.show()
```

```python
# pie plot
slices = [7,2,2,13]
activities = ['Sleeping','Eating','Playing','Working']
cols = ['c','m','r','y']
plt.pie(slices,
        labels = activities,
        colors = cols,
        startangle = 90,
        shadow = True,
        explode = (0,0.1,0,0),
        autopct = '%1.1f%%') # overlay percentage over graph
plt.show()
```

```python
# how to change the lables position
from matplotlib import style
style.use('seaborn')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'g',label='Line One', linewidth=5) # 'g' = green
plt.plot(x2,y2,'c', label='Line Two', linewidth=5)
plt.title('Info')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend(loc=3)
plt.grid(True,color='k') # k --> black color
plt.show()
```
