```python
# Exploratory Data Analysis with Pandas

# Use Telecom Industry data - churn users(who left the subscription)

import numpy as np
import pandas as pd
```

```python
# Load the Dataset
df = pd.read_csv('https://trello-attachments.s3.amazonaws.com/600d1f10d700af20b1924b3c/600d1f700fcd073384905133/4875f4c4af388917d8508e224b3d7219/telecom_churn.csv')
df.head()
```

```python
# columns --> retrieve the column names
df.columns
```

```python
df.shape
```

```python
# info() --> used to generate the full summary of dataset
df.info()
```

```python
# astype() --> used to change the dataType
df['churn'] = df['churn'].astype('int64')
df.info()
```

```python
# describe()
# used to statistical summary of dataset
# by default this method works only on int and float values

df.describe()
```

```python
# in order to see the statistical summary of non numerical features
# use include parameter in describe()

df.describe(include=['object','bool'])
```

```python
data = pd.Series(['A','B','C','A','C','D','E','C','B','C','B','A'])
data.describe()
```

```python
# value_counts()
df['churn'].value_counts()
```

```python
df.head(3)
```

```python
# sorting
df.sort_values(by='total day charge').head(10) # ascending order
```

```python
# descending order sorting
df.sort_values(by='total day charge', ascending=False).head(10)
```

```python
# maximum value of total day charge
df['total day charge'].max()
```

```python
# how much time on an average do churn users spent on phone during
# day time ??

df[df['churn']==1]['total day minutes'].mean()
```

```python
df[df['churn']==0]['total day minutes'].mean()
```

```python
# what is the maximum length of international calls among loyal users
df[df['churn']==0]['total intl calls'].max()
```

```python
df.max()
```

```python
# retrieve only those records where state starts with the letter 'w'

df[df['state'].apply(lambda state: state[0]=='W')].head()
```
