```python
from bs4 import BeautifulSoup
webpage = requests.get('https://fortifysecurity.org/')
# Create BeautifulSoup Object
soup = BeautifulSoup(webpage.content,'html.parser')
```

```python
# scrape title of webpage
page_title = soup.title
print(page_title)
```

```python
page_title = soup.title.text
print(page_title)
```

```python
# Extract Head Section of Webpage
page_head = soup.head
print(page_head)
```

```python
# Extract Body of Webpage
page_body = soup.body
print(page_body)
```

```python
# BeautifulSoup Methods
# select()
# used to select all the instances of specified HTML tag
# it returns Python List

h1_list = soup.select('h1')
print(h1_list)
print(len(h1_list))
```

```python
# Extract text of first H1 Tag
first_h1 = soup.select('h1')[0].text
print(first_h1)
```

```python
# Task
# Select the text of all H1 Tags , and then we need to store
# these text in a new list

all_h1_tags = []

for everyh1 in soup.select('h1'):
    all_h1_tags.append(everyh1.text)
    
print(all_h1_tags)
```

```python
# find_all()
# used to return a list of all instances of specified HTML tag

soup.find_all('img')
```

```python
# fetching SRC

images = soup.find_all('img')

for everyImage in images:
    print('https://www.fortifysecurity.org/'+everyImage['src'])
```

```python
# Task Number 1
# Fetch the HREF information of all anchor 'a' tag
# Also print the number of anchor tags 
#  Output : Number Of Anchor Tags : 60


# Task Number 2
# Store image 'src' and 'alt' attribute information in Pandas Data Frame
# Output

#  src                                       alt
# admin/course_image/crime_defender.jpg     Crime Defender
# if alt text is not available you need to replace it with Not Available
# assets/img/about.jpg                      Not Available
```

```python
# # Task Number 1

href_list = soup.select('a')
    
for everyhref in href_list:
    print(everyhref)

print("Number of Anchor Tags :",len(href_list))
```

```python
# Task 2
import pandas as pd

images = soup.find_all('img')


src_data = []
alt_data = []

for everydata in images:

    src_data.append(everydata['src'])
    alt_data.append(everydata['alt']) 
    
images_data = {'src': src_data, 
               'alt': alt_data}

data = pd.DataFrame(images_data)

a=0
while a<len(data):
    if len(data.alt[a])==0:
        print(data.src[a])
        data.alt[a] = 'Not Available'
    a=a+1
    
data
```
