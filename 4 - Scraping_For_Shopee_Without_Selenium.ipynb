```python
# WEB SCRAPING FOR SHOPEE

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# generate url for products

def get_url(product_name):
    product_name = product_name.replace(' ','%20')
    template = 'https://shopee.com.my/search?keyword={}'
    url = template.format(product_name)
    return url
    
#product_name = product_name.replace("'s",'%27')
```

```python
url = get_url('jeans')
print(url)
```

```python
# check response

response = requests.get(url)
print(response)
```

```python
# create BeautifulSoup object

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# search for all div tags having class name col-xs-2-4

cards = soup.find_all('div','col-xs-2-4 shopee-search-item-result__item')
print(len(cards))
```
