```python
# Project 1 : Guess A Number Game
# Our computer will generate one random number
# We need to predict that number

import random
guessAttempts = 0
print('Hello! What is your name ?')
myName = input()
number = random.randint(1,13)
print('Well', myName, 'I am thinking of a number between 1 to 14. Please Wait..')

while guessAttempts < 4:
  print('Take a guess : ')
  guess = input()
  guess = int(guess)

  guessAttempts = guessAttempts + 1

  if guess > number:
    print('Your Guess Is Too HIGH')

  if guess < number:
    print('Your Guess Is Too LOW')
  
  if guess == number:
    break

if guess == number:
  print('Great Job', myName, 'You Guesses The Number In', guessAttempts, 'Attempts')

if guess != number:
  print('Game Over ! The Number I was Thinking Is : ', number)
```
