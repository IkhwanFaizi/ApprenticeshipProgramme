```python
# Conditional Statements / Working With Condition / Decision Making
# if   elif   else
# For Loop

# Decision making is required when we want to execute a code only 
# if a certain condition is satisfied.

# Python if Statement is used for decision-making operations. 
# It contains a body of code which runs only when the condition given 
# in the if statement is true.

# Syntax

# if condition/test_expression:
#    python code
```

```python
age = 15

if age>=15:
  print('You Can Play The Game')
```

```python
num = int(input('Enter Any Number : '))

if num%2==0:
  print('Checking Number.....')
  print('Even Number')
```

```python
# Task
# Declare 5 int variables
# We need to determine which one is the largest

a = 19
b = 45
c = 27
d = 148
e = 5600

# Output : Largest is 5600
# Do not use logical operators [ AND or NOT]

num = 0

if num < a:  # or if a>num
  num = a

if num < b:
  num = b

if num < c:
  num = c

if num < d:
  num = d

if num < e:
  num = e

print('The largest is', num)
```

```python
# if else

age = 11

if age>12:
  print('Game is Loading !')
else:
  print('Age Restriction, Age Should Be Above 12 Years Old')
```

```python
# if elif else

num = int(input('Enter Any Number : '))

if num<0:
  print('Negative Number')
elif num==0:
  print('Number Is Zero')
else:
  print('Positive Number')    
```

```python
marks = int(input('Enter Any Number : '))

if marks<=100 and marks>=90:
  print('A Grade')
elif marks<=89 and marks>=75:
  print('B Grade')
elif marks<=74 and marks>=65:
  print('C Grade')
elif marks<=64 and marks>=50:
  print('D Grade')
else:
  print('Fail') 
```
